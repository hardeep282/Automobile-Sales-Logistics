```python
#  Basic cleaning (drop + date convert)


import pandas as pd
import numpy as np

# Copy original
df = data.copy()

# 1) days_since_lastorder drop 
if 'days_since_lastorder' in df.columns:
    df = df.drop(columns=['days_since_lastorder'])

# 2) orderdate to datetime conversion
df['orderdate'] = pd.to_datetime(df['orderdate'], dayfirst=True)

# 3) Analysis date for RFM
analysis_date = df['orderdate'].max() + pd.Timedelta(days=1)
print("Analysis date:", analysis_date.date())
```

```python
# STEP 1 – Customer-level RFM + extra features

# Helper: mode safely lene ke liye
def safe_mode(x):
    m = x.mode()
    if m.empty:
        return np.nan
    return m.iloc[0]

# 👉 Customer-level aggregation
cust = (
    df
    .groupby('customername')
    .agg(
        recency_days = ('orderdate', lambda x: (analysis_date - x.max()).days),
        frequency_orders = ('ordernumber', 'nunique'),
        monetary_sales = ('sales', 'sum'),
        
        avg_totalmsrp = ('totalmsrp', 'mean'),
        n_unique_products = ('productcode', 'nunique'),
        
        # Shipping reliability = % of orders shipped
        shipping_reliability = ('shipping_status', lambda x: (x == 'Shipped').mean()),  # Did not understood this shipping reliability 1 how ?
        
        # Dominant attributes per customer
        main_country = ('country', safe_mode),
        main_city = ('city', safe_mode),
        main_productline = ('productline', safe_mode),
        main_dealsize = ('dealsize', safe_mode)
    )
    .reset_index()
)

print(cust.head())
print(cust.shape)
```
